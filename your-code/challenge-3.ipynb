## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: Aggregate the amount_spent for unique customers
customer_expenses = orders.groupby('CustomerID')['amount_spent'].sum()
customer_expenses.head()
```

```python
# Sub Problem 2: Select customers whose aggregated amount_spent is in a given quantile range
vip_quantile = customer_expenses.quantile(0.95)
vip_quantile
```

```python
preferred_quantile = customer_expenses.quantile(0.75)
preferred_quantile
```

```python
# Sub Problem 3: Label selected customers as "VIP" or "Preferred"
vip_customers = customer_expenses[customer_expenses > vip_quantile].index.tolist()
vip_customers
```

```python
preferred_customers = customer_expenses[(customer_expenses > preferred_quantile) & 
                                        (customer_expenses <= vip_quantile)].index.tolist()
preferred_customers
```

```python
def classify_customer(amount_spent):
    if amount_spent > vip_quantile:
        return 'VIP'
    elif amount_spent > preferred_quantile:
        return 'Preferred'
    else:
        return 'Regular'
```

```python
customer_classification = customer_expenses.apply(classify_customer)
customer_classification.value_counts() 
```

```python
customer_classification = customer_classification.rename("Customer_Classification")
```

```python
orders_with_classification = orders.merge(customer_classification, left_on='CustomerID', right_index=True, how='left')
orders_with_classification
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# Filter orders by Customer_Classification = VIP
# Group by Country
# Order by number of unique customers
# Limit 1

country_max_vip = (
    orders_with_classification[orders_with_classification['Customer_Classification']=='VIP']
    .groupby('Country')['CustomerID']
    .nunique()
    .sort_values(ascending=False)
    .head(1)
)
country_max_vip

print(f'{country_max_vip.index[0]} has the most VIP Customers with total number of {country_max_vip.values[0]} customers')
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
country_max_vip_preferred = (
    orders_with_classification[(orders_with_classification['Customer_Classification']=='VIP') | 
                              (orders_with_classification['Customer_Classification']=='Preferred')]
    .groupby('Country')['CustomerID']
    .nunique()
    .sort_values(ascending=False)
    .head(1)
)
country_max_vip_preferred

print(f'{country_max_vip_preferred.index[0]} has the most VIP+Preferred Customers with total number of {country_max_vip_preferred.values[0]} customers')
```
